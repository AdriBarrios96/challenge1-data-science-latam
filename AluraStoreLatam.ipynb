# 4. Productos más y menos vendidos

###En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
#Libreria de Graficos
import matplotlib.pyplot as plt

#Calculamos las ventas de cada producto
ventas_tienda = tienda['Producto'].value_counts()

print("- - - Ventas de tienda 1 - - - \n")
print("Los productos MAS vendidos son \n", ventas_tienda.head(3))
print("\nLos productos MENOS vendidos son \n", ventas_tienda.tail(3))
```

```python
#Grafico de Barras
ventas_tienda.plot(kind='barh' , figsize=(20, 18), color='red')

plt.title('Ventas de Productos - Tienda 1')
plt.xlabel('Productos')
plt.ylabel('Cantidad de Ventas')
```

# GRÁFICO 3: DISPERSIÓN (Precio vs Costo de Envío)
####Veremos si existe un patrón

```python
plt.figure(figsize=(9, 5))
# plt.scatter crea un punto por cada fila que cruza Precio (X) y Envío (Y)
plt.scatter(tienda['Precio'], tienda['Costo de envío'], color='green', alpha=0.9)
plt.title('Relación: Precio del Producto vs Costo de Envío (Tienda 1)', fontsize=14, fontweight='bold')
plt.xlabel('Precio del Producto ($)', fontsize=12)
plt.ylabel('Costo de Envío ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5) # Le agregamos una cuadrícula de fondo
plt.show()
```
